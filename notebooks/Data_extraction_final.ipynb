```python
def data_extraction(year, path):
    
    ##list of target drug names
    target_drugs = ['XENICAL', 'ALLI', 'ORLISTAT']
    
    ##path to q1 ascii files
    q1_path = path + str(year)+"/faers_ascii_"+str(year)+"q1" + "/ASCII/"

    ##read demo data
    q1_DEMO = pd.read_csv(q1_path+f'DEMO{year%100}Q1.txt', delimiter = '$', dtype = str)

    ##read drug data
    q1_DRUG = pd.read_csv(q1_path+f'DRUG{year%100}Q1.txt', delimiter = '$', dtype = str)

    ##read indication data
    q1_INDI = pd.read_csv(q1_path+f'INDI{year%100}Q1.txt', delimiter = '$', dtype = str)

    ##read outcome data
    q1_OUTC = pd.read_csv(q1_path+f'OUTC{year%100}Q1.txt', delimiter = '$', dtype = str)

    ##read reaction data
    q1_REAC = pd.read_csv(q1_path+f'REAC{year%100}Q1.txt', delimiter = '$', dtype = str)

    ##filter the target drugs
    q1_DRUG_filtered = q1_DRUG.loc[q1_DRUG['drugname'].isin(target_drugs)]

    ##quarterly data list
    #q1_dfs = [q1_DEMO, q1_DRUG, q1_INDI, q1_OUTC, q1_REAC]

    ##merge DRUG data and DEMO data
    merged_df_1 = q1_DRUG_filtered.merge(q1_DEMO, on = ['primaryid', 'caseid'], how = 'inner')

    ##merge the indication data
    merged_df_1 = merged_df_1.merge(q1_INDI, on = ['primaryid', 'caseid'], how = 'inner')

    ##merge the outcome data
    merged_df_1 = merged_df_1.merge(q1_OUTC, on = ['primaryid', 'caseid'], how = 'inner')

    ##merge the reaction data
    merged_df_1 = merged_df_1.merge(q1_REAC, on = ['primaryid', 'caseid'], how = 'inner')

    
    ##path to q2 ascii files
    q2_path = path + str(year)+"/faers_ascii_"+str(year)+"q2" + "/ASCII/"

    ##read demo data
    q2_DEMO = pd.read_csv(q2_path+f'DEMO{year%100}Q2.txt', delimiter = '$', dtype = str)

    ##read drug data
    q2_DRUG = pd.read_csv(q2_path+f'DRUG{year%100}Q2.txt', delimiter = '$', dtype = str)

    ##read indication data
    q2_INDI = pd.read_csv(q2_path+f'INDI{year%100}Q2.txt', delimiter = '$', dtype = str)

    ##read outcome data
    q2_OUTC = pd.read_csv(q2_path+f'OUTC{year%100}Q2.txt', delimiter = '$', dtype = str)

    ##read reaction data
    q2_REAC = pd.read_csv(q2_path+f'REAC{year%100}Q2.txt', delimiter = '$', dtype = str)

    ##filter the target drugs
    q2_DRUG_filtered = q2_DRUG.loc[q2_DRUG['drugname'].isin(target_drugs)]

    ##quarterly data list
    #q1_dfs = [q1_DEMO, q1_DRUG, q1_INDI, q1_OUTC, q1_REAC]

    ##merge DRUG data and DEMO data
    merged_df_2 = q2_DRUG_filtered.merge(q2_DEMO, on = ['primaryid', 'caseid'], how = 'inner')

    ##merge the indication data
    merged_df_2 = merged_df_2.merge(q2_INDI, on = ['primaryid', 'caseid'], how = 'inner')

    ##merge the outcome data
    merged_df_2 = merged_df_2.merge(q2_OUTC, on = ['primaryid', 'caseid'], how = 'inner')

    ##merge the reaction data
    merged_df_2 = merged_df_2.merge(q2_REAC, on = ['primaryid', 'caseid'], how = 'inner')

    
    ##path to q3 ascii files
    q3_path = path + str(year)+"/faers_ascii_"+str(year)+"q3" + "/ASCII/"

    ##read demo data
    q3_DEMO = pd.read_csv(q3_path+f'DEMO{year%100}Q3.txt', delimiter = '$', dtype = str)

    ##read drug data
    q3_DRUG = pd.read_csv(q3_path+f'DRUG{year%100}Q3.txt', delimiter = '$', dtype = str)

    ##read indication data
    q3_INDI = pd.read_csv(q3_path+f'INDI{year%100}Q3.txt', delimiter = '$', dtype = str)

    ##read outcome data
    q3_OUTC = pd.read_csv(q3_path+f'OUTC{year%100}Q3.txt', delimiter = '$', dtype = str)

    ##read reaction data
    q3_REAC = pd.read_csv(q3_path+f'REAC{year%100}Q3.txt', delimiter = '$', dtype = str)

    ##filter the target drugs
    q3_DRUG_filtered = q3_DRUG.loc[q3_DRUG['drugname'].isin(target_drugs)]

    ##quarterly data list
    #q1_dfs = [q1_DEMO, q1_DRUG, q1_INDI, q1_OUTC, q1_REAC]

    ##merge DRUG data and DEMO data
    merged_df_3 = q3_DRUG_filtered.merge(q3_DEMO, on = ['primaryid', 'caseid'], how = 'inner')

    ##merge the indication data
    merged_df_3 = merged_df_3.merge(q3_INDI, on = ['primaryid', 'caseid'], how = 'inner')

    ##merge the outcome data
    merged_df_3 = merged_df_3.merge(q3_OUTC, on = ['primaryid', 'caseid'], how = 'inner')

    ##merge the reaction data
    merged_df_3 = merged_df_3.merge(q3_REAC, on = ['primaryid', 'caseid'], how = 'inner')

    
    ##path to q4 ascii files
    q4_path = path + str(year)+"/faers_ascii_"+str(year)+"Q4" + "/ASCII/"

    ##read demo data
    q4_DEMO = pd.read_csv(q4_path+f'DEMO{year%100}Q4.txt', delimiter = '$', dtype = str)

    ##read drug data
    q4_DRUG = pd.read_csv(q4_path+f'DRUG{year%100}Q4.txt', delimiter = '$', dtype = str)

    ##read indication data
    q4_INDI = pd.read_csv(q4_path+f'INDI{year%100}Q4.txt', delimiter = '$', dtype = str)

    ##read outcome data
    q4_OUTC = pd.read_csv(q4_path+f'OUTC{year%100}Q4.txt', delimiter = '$', dtype = str)

    ##read reaction data
    q4_REAC = pd.read_csv(q4_path+f'REAC{year%100}Q4.txt', delimiter = '$', dtype = str)

    ##filter the target drugs
    q4_DRUG_filtered = q4_DRUG.loc[q4_DRUG['drugname'].isin(target_drugs)]

    ##quarterly data list
    #q1_dfs = [q1_DEMO, q1_DRUG, q1_INDI, q1_OUTC, q1_REAC]

    ##merge DRUG data and DEMO data
    merged_df_4 = q4_DRUG_filtered.merge(q4_DEMO, on = ['primaryid', 'caseid'], how = 'inner')

    ##merge the indication data
    merged_df_4 = merged_df_4.merge(q4_INDI, on = ['primaryid', 'caseid'], how = 'inner')

    ##merge the outcome data
    merged_df_4 = merged_df_4.merge(q4_OUTC, on = ['primaryid', 'caseid'], how = 'inner')

    ##merge the reaction data
    merged_df_4 = merged_df_4.merge(q4_REAC, on = ['primaryid', 'caseid'], how = 'inner')
    
    return [merged_df_1, merged_df_2, merged_df_3, merged_df_4]
```

```python
##convert age to same unit -> year

##mapping function
def convert_age(row):
    
    if row['age_cod'] == 'YR': ##age is in years
        
        ##return yrs
        return row['age']
    
    elif row['age_cod'] == 'MON': ##age is in months
        
        ##convert months to yrs
        return row['age']/12
    
    elif row['age_cod'] == 'DY': ##age is in days
        
        ##convert days to yrs
        return row['age']/365
    
    elif row['age_cod'] == 'DEC': ##age is in decade
        
        ##convert decade to yrs
        return row['age']*10
        
    else:
        return None
    
##apply the mapping function to df
five_yrs_data_cpy['age_in_yrs'] = five_yrs_data_cpy.apply(convert_age, axis = 1)
```

```python
##convert weight to same unit -> kg

##mapping function
def convert_weight(row):
    
    if row['wt_cod'] == 'KG': ##weight is in kgs
        
        ##return weight
        return row['wt']
    
    elif row['wt_cod'] == 'LBS': ##weight is in lbs
        
        ##convert weight to kgs
        return row['wt']*0.45359237
        
    else:
        return row['wt']
    
##apply the mapping function to df
five_yrs_data_cpy['wt_in_kgs'] = five_yrs_data_cpy.apply(convert_weight, axis = 1)
```
